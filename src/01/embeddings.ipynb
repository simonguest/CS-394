```python
# Load the sentence tranformer model

from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Show the embeddings for a single sentence

embeddings = model.encode("I love pizza!")
embeddings
```

```python
# Compare the similarity of three sentences

from sklearn.metrics.pairwise import cosine_similarity

sentences = ['The cat sat on the mat', 
             'The dog rested on the rug',
             'I love pizza']
embeddings = model.encode(sentences)

print(cosine_similarity(embeddings))
```
